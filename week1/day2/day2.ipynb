## 3. Functions
Writing reusable functions with parameters, return values, and docstrings.

```python
def normalize(value, min_val , max_val):
    """scale a value to the 0-1 range."""
    return (value - min_val) / (max_val - min_val)
print(normalize(75 , 0 , 100))
print(normalize(50, 0 ,200))
```

## 5. Object-Oriented Programming (OOP) Basics
A class bundles data (attributes) and behavior (methods) together.

```python
class DataPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5

p = DataPoint(3, 4)
print(p.distance_from_origin())   
```

## Hands-On Lab: Python Fundamentals Drill

### Step 1: Mean, Min, Max as a Dictionary

```python
def summarize(numbers):
    """Return the mean, min, and max of a list of numbers as a dictionary."""
    avg = sum(numbers) / len(numbers)            
    smallest = min(numbers)           
    largest = max(numbers)          
    
    return {
        "mean": avg,
        "min": smallest,
        "max": largest
    }
print(summarize([4, 8, 15, 16, 23, 42]))
```

This function takes a list of numbers and returns a dictionary with the mean, min, and max values — useful for quick summary statistics.

### Step 2: Rewrite a Loop as a List Comprehension

```python
# النسخة الأصلية (loop):
evens_loop = []
for x in range(20):
    if x % 2 == 0:
        evens_loop.append(x)
evens = [x for x in range(20) if x % 2 == 0]
print(evens)
```

Rewrote the loop-based even-number filter as a single-line list comprehension — more concise and idiomatic Python than the equivalent for-loop.

### Step 3: A Small Class Representing a Data Record

```python
class Student:
    def __init__(self, name, grades):
        self.name = name        
        self.grades = grades    

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

s1 = Student("Sara", [85, 90, 78])
print(s1.name)
print(s1.average_grade())
```

Defined a `Student` class with `name` and `grades` as attributes, and `average_grade` as a method that computes the mean grade — a simple example of bundling data and behavior together (OOP).

## 🌟 Bonus: Extended Summary Statistics
Extending the `summarize` function from the Hands-On Lab to also compute the median and standard deviation, without using external libraries.

```python
def median(numbers):
    sorted_numbers = sorted(numbers)    
    n = len(sorted_numbers)         
    
    if n % 2 == 0:
        mid1 = sorted_numbers[n//2 - 1]
        mid2 = sorted_numbers[n//2]
        return (mid1 + mid2) / 2
    else:
        return sorted_numbers[n//2]  
print(median([7, 1, 3]))   
print(median([1, 2, 3, 4]))       
```

```python
def std_dev(numbers):
    mean = sum(numbers) / len(numbers)
    
    squared_diffs = [(x - mean)**2 for x in numbers]  
    variance = sum(squared_diffs) / len(squared_diffs)
    
    return variance ** 0.5              

print(std_dev([2, 4, 4, 4, 5, 5, 7, 9]))  
```

Implemented `median()` and `std_dev()` from scratch (without NumPy) to extend the original `summarize` function. Tested both with sample data to confirm correct results (median=3/2.5, std_dev=2.0).

```python
def summarize_extended(numbers):
    """Extended version: adds median and standard deviation to the original summary."""
    basic = summarize(numbers)  
    basic["median"] = median(numbers)
    basic["std_dev"] = std_dev(numbers)
    return basic

print(summarize_extended([2, 4, 4, 4, 5, 5, 7, 9]))
```
