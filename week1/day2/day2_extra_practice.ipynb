## 1️⃣ Median from Scratch

Calculating the median manually: sorting the numbers, then picking the middle value (or averaging the two middle values if the list has an even length).

```python
def median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)

    if n % 2 == 0:
        mid1 = sorted_numbers[n//2 - 1]
        mid2 = sorted_numbers[n//2]
        return (mid1 + mid2) / 2
    else:
        return sorted_numbers[n//2]

print(median([7, 1, 3]))
print(median([1, 2, 3, 4]))
```

## 2️⃣ Standard Deviation from Scratch

Measuring how spread out the numbers are from their mean — computed step by step (mean → squared differences → variance → square root), without NumPy.

```python
def std_dev(numbers):
    mean = sum(numbers) / len(numbers)

    squared_diffs = [(x - mean)**2 for x in numbers]
    variance = sum(squared_diffs) / len(squared_diffs)

    return variance ** 0.5

print(std_dev([2, 4, 4, 4, 5, 5, 7, 9]))
```

## 0️⃣ Base Function (Recap)

Re-defining the original `summarize()` function from Day 2's Hands-On Lab, since this notebook builds on top of it.

```python
def summarize(numbers):
    """Return the mean, min, and max of a list of numbers as a dictionary."""
    avg = sum(numbers) / len(numbers)
    smallest = min(numbers)
    largest = max(numbers)

    return {
        "mean": avg,
        "min": smallest,
        "max": largest
    }
def summarize_extended(numbers):
    """Extended version: adds median and standard deviation to the original summary."""
    basic = summarize(numbers)
    basic["median"] = median(numbers)
    basic["std_dev"] = std_dev(numbers)
    return basic

print(summarize_extended([2, 4, 4, 4, 5, 5, 7, 9]))
```
