## 📥 1. Load and Clean the Data

Reusing the cleaning approach from Day 4: dropping `Cabin` (too many missing values),
filling missing `Age` with the column mean, and dropping the few rows missing `Embarked`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

df = df.drop(columns=["Cabin"])
df["Age"] = df["Age"].fillna(df["Age"].mean())
df = df.dropna(subset=["Embarked"])

df.head()
```

## 🔢 2. Derived Statistic with NumPy

Using NumPy to compute the overall survival rate as a percentage — a single derived
numeric feature that summarizes the outcome of the voyage at a glance.

```python
survival_rate = np.mean(df["Survived"]) * 100
print(f"Overall survival rate: {survival_rate:.2f}%")
```

## 📊 3. Visualizing the Data

### 3.1 Age Distribution (Histogram)
Shows the overall spread of passenger ages, revealing whether the ship carried
mostly young adults, families, or a broad mix.

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Age"], bins=20, color="steelblue", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

### 3.2 Survival Count by Passenger Class (Bar Chart)
Compares the number of survivors across the three passenger classes, highlighting
whether class was associated with survival chances.

```python
survivors_by_class = df[df["Survived"] == 1]["Pclass"].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(survivors_by_class.index.astype(str), survivors_by_class.values, color="seagreen")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Survivors")
plt.title("Survivors by Passenger Class")
plt.show()
```

### 3.3 Age vs. Fare (Scatter Plot)
Explores the relationship between passenger age and ticket fare, to check whether
older or younger passengers tended to pay more.

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Age"], df["Fare"], alpha=0.5, color="darkorange")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs. Fare")
plt.show()
```
