### Exercise C: Shopping Bill

Given arrays of prices and quantities for different products, we compute 
the cost per item, the total bill, and apply a discount — all using 
vectorized operations.

```python
prices = np.array([10, 25, 5, 40])
quantities = np.array([2, 1, 10, 3])
item_costs = prices * quantities
print("Cost per item:", item_costs)
total_bill = item_costs.sum()
print("Total bill:", total_bill)
discount = 0.10
final_price = total_bill * (1 - discount)
print("Final price after 10% discount:", final_price)
```
