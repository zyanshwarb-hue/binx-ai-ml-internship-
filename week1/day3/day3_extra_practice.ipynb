Creating arrays using three different NumPy methods:

| Method | Purpose |
|--------|---------|

| `zeros` | A blank array pre-filled with zeros |
| `linspace` | Evenly spaced values between two numbers |

```python
import numpy as np
arr = np.random.randint(1, 101, size= 10)
print(arr)
```

```python
import numpy as np
arr = np.zeros((3, 5))
print(arr)
```

### 🅲️ Exercise C — Shopping Bill

**Scenario:** Arrays hold prices and quantities for different products.

**Tasks:**
1. Compute the cost per item (`prices * quantities`)
2. Calculate the total bill
3. Apply a 10% discount — all using vectorized operations, no loops

```python
prices = np.array([10, 25, 5, 40])
quantities = np.array([2, 1, 10, 3])
item_costs = prices * quantities
print("Cost per item:", item_costs)
total_bill = item_costs.sum()
print("Total bill:", total_bill)
discount = 0.10
final_price = total_bill * (1 - discount)
print("Final price after 10% discount:", final_price)
```

**Tasks:**
1. Increase brightness by adding a constant value to every pixel

3. Invert the image by subtracting from 255

```python
image = np.array([[100, 150, 200],
                   [50, 75, 125],
                   [10, 20, 30]])
Brilliant_Image = image +30
print(Brilliant_Image)
clipped_image = np.clip(Brilliant_Image,0,255)
The_inverse_matrix = 255 - clipped_image 
print(The_inverse_matrix)
```
