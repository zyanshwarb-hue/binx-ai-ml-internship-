## Objective
Practicing array creation, indexing, slicing, boolean masking, and broadcasting using **NumPy**.

## Step 1: Array Creation

We create a `4x4` 2D array containing values from 1 to 16 using `np.arange()` and `.reshape()`, 
then inspect its **shape** and **dtype**.

```python
import numpy as np
arr = np.arange(1,17).reshape(4,4)
print(arr)
print("Shape:", arr.shape)
print("Dtype:", arr.dtype)
```

## Step 2: Indexing & Slicing

We extract the **second column** (`arr[:, 1]`) and the **last row** (`arr[-1, :]`) 
using NumPy's slicing syntax.

```python
second_column = arr[:, 1]
last_row = arr[-1, :]

print("Second column:", second_column)
print("Last row:", last_row)
```

## Step 3: Boolean Masking

We compute the mean of the array, then use a **boolean mask** (`arr > mean`) 
to select only the values greater than the mean.

```python
mean_value = arr.mean()
above_mean = arr[arr > mean_value]

print("Mean value:", mean_value)
print("Values above mean:", above_mean)
```

## Step 4: Broadcasting

We add a 1D array (`row_to_add`) to every row of the 2D array using **broadcasting**, 
then manually verify the result for the first row.

```python
row_to_add = np.array([100, 200, 300, 400])
result = arr + row_to_add

print("Original array:\n", arr)
print("\nRow to add:", row_to_add)
print("\nResult after broadcasting:\n", result)
```

### Manual Verification
We manually compute the broadcasting result for row 0 and confirm it matches NumPy's output.

```python
manual_check = arr[0] + row_to_add
print("Manual check for row 0:", manual_check)
print("Matches broadcasting result?", np.array_equal(manual_check, result[0]))
```
